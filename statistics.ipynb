````python
import json
    
exps = []
with open ('SCORE.txt', 'r', encoding='utf-8') as f:
    rs = f.read()
    rs = rs.split('========== Lần lặp thứ ==========')
    rs = rs[1:]
    
    for i, exp in enumerate(rs):
        exp = exp.split("\n\n")
        exp = exp[:-1]

        temp = []
        for j, qa in enumerate(exp):
            qa_clean = qa.strip()

            if qa_clean.startswith("```json"):
                qa_clean = qa_clean[len("```json"):].strip()
            if qa_clean.endswith("```"):
                qa_clean = qa_clean[:-3].strip()

            qa_clean = qa_clean.replace("'", '"')
            qa_dict = json.loads(qa_clean)
            
            temp.append(qa_dict)
        exps.append(temp)

# for i, exp in enumerate(exps):
#     print(i, ":", len(exp))
````

````python
import pandas as pd

category_mapping = {
    'Triển khai':              'Deployment',
    'Chủ đề Nâng cao':         'Advanced',
    'Câu hỏi bổ sung':         'Extras',
    'Hiểu biết chung về ISOC': 'ISOC Basics',
    'Lập kế hoạch và Thiết kế':'Plan & Design',
    'Thành phần và Công nghệ': 'Components & Tech',
    'Vận hành và Quản lý':     'Operations & Management',
}

all_data = []
for i, exp in enumerate(exps, 1):
    df = pd.DataFrame(exp)
    df['experiment_id'] = i
    all_data.append(df)

df_all = pd.concat(all_data, ignore_index=True)
summary_category = df_all.groupby('category').agg(
    accuracy_mean=('accuracy', 'mean'),
    accuracy_std=('accuracy', 'std'),
    fluency_mean=('fluency', 'mean'),
    fluency_std=('fluency', 'std'),
    relevance_mean=('relevance', 'mean'),
    relevance_std=('relevance', 'std'),
).reset_index()

summary_category['Category'] = summary_category['category'].map(category_mapping)
cols = ['Category'] + [col for col in summary_category.columns if col not in ['category', 'Category']]
summary_category = summary_category[cols].round(3)

def mean_std_format(mean, std):
    return f"{mean:.3f} ± {std:.3f}"

summary_category['Accuracy'] = summary_category.apply(
    lambda row: mean_std_format(row['accuracy_mean'], row['accuracy_std']),
    axis=1
)
summary_category['Fluency'] = summary_category.apply(
    lambda row: mean_std_format(row['fluency_mean'], row['fluency_std']),
    axis=1
)
summary_category['Relevance'] = summary_category.apply(
    lambda row: mean_std_format(row['relevance_mean'], row['relevance_std']),
    axis=1
)

# Chọn lại cột trình bày cuối cùng
summary_category = summary_category[['Category', 'Accuracy', 'Fluency', 'Relevance']]

print("\n=== Thống kê theo category (dạng mean ± std) ===")
print(summary_category)
````

````python
import pandas as pd

def summarize_experiment(exp):
    df = pd.DataFrame(exp)
    overall = df[['accuracy', 'fluency', 'relevance']].agg(['count', 'mean', 'std'])
    by_category = df.groupby('category')[['accuracy', 'fluency', 'relevance']].agg(['count', 'mean', 'std'])
    return overall.round(3), by_category.round(3)

summary_rows = []
for idx, exp in enumerate(exps, start=1):
    overall, _ = summarize_experiment(exp)
    row = {
        'Experiment': f'Exp {idx}',
        'Accuracy': f"{overall.loc['mean', 'accuracy']} ± {overall.loc['std', 'accuracy']}",
        'Fluency': f"{overall.loc['mean', 'fluency']} ± {overall.loc['std', 'fluency']}",
        'Relevance': f"{overall.loc['mean', 'relevance']} ± {overall.loc['std', 'relevance']}",
    }
    summary_rows.append(row)

summary_df = pd.DataFrame(summary_rows)
print(summary_df)
````

````python
def categorize_accuracy(value):
    if value > 0.60:
        return 'High'
    elif value >= 0.50:
        return 'Medium'
    else:
        return 'Low'

def summarize_by_accuracy_group(df):
    # Tính Accuracy trung bình cho mỗi hàng nếu cần
    df['accuracy_mean'] = df['Accuracy'].apply(lambda x: float(str(x).split('±')[0].strip()))
    
    # Gán nhóm hiệu suất
    df['Performance Group'] = df['accuracy_mean'].apply(categorize_accuracy)
    
    # Đếm số lượng theo nhóm
    group_counts = df['Performance Group'].value_counts().reset_index()
    group_counts.columns = ['Performance Group', 'Number of Experiments']
    
    return group_counts

group_summary = summarize_by_accuracy_group(summary_df)
print(group_summary)
````

````python
import pandas as pd

def summarize_experiment(exp):
    df = pd.DataFrame(exp)
    overall = df[['accuracy', 'fluency', 'relevance']].agg(['count', 'mean', 'std'])
    by_category = df.groupby('category')[['accuracy', 'fluency', 'relevance']].agg(['count', 'mean', 'std'])
    return overall.round(3), by_category.round(3)


category_mapping = {
    'Triển khai':              'Deployment',
    'Chủ đề Nâng cao':         'Advanced',
    'Câu hỏi bổ sung':         'Extras',
    'Hiểu biết chung về ISOC': 'ISOC Basics',
    'Lập kế hoạch và Thiết kế':'Plan & Design',
    'Thành phần và Công nghệ': 'Components & Tech',
    'Vận hành và Quản lý':     'Operations & Management',
}

metrics = ['accuracy', 'fluency', 'relevance']
experiments_data = []

for idx, exp in enumerate(exps, start=1):
    _, by_cat = summarize_experiment(exp)
    df = by_cat[[(m, 'mean') for m in metrics]].copy()
    df.columns = metrics
    df['Experiment'] = f'Exp {idx}'
    
    df['Category'] = df.index.map(category_mapping)
    experiments_data.append(df.reset_index(drop=True))

all_data = pd.concat(experiments_data, ignore_index=True)
all_data_grouped = all_data.groupby(['Category', 'Experiment'], as_index=False).mean()

for metric in metrics:
    print(f"\n=== {metric.capitalize()} by Category and Experiment ===")
    metric_table = all_data_grouped.pivot(index='Category', columns='Experiment', values=metric)
    print(metric_table)
````
